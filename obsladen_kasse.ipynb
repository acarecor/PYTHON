```python

aepfel_lager = 15
birnen_lager= 20
banananen_lager= 10

bestell_aepfel= 0
bestell_birnen= 0
bestell_bananen= 0

preis_aepfel= 0.8
preis_birnen= 0.5
preis_bananen= 0.2

summe = 0.0
geld = 0.0
rueckgeld = 0.0

print('Kasse Obsladen')
print('-------------------------------')

bestell_aepfel = int(input('Wieviel Äpfel hätten sie gern '))
summe_aepfel = bestell_aepfel * preis_aepfel
if bestell_aepfel <= bestell_aepfel:
    print(bestell_aepfel, 'Äpfel                       EUR', "%.2f" % summe_aepfel)
    
else: 
    bestell_aepfel > aepfel_lager
    print('Tut mir Leid es gibt nich genug Äpfel in Lager')
     

bestell_birnen = int(input('Wieviel Birnen hätten sie gern'))
summe_birnen = bestell_birnen * preis_birnen
if bestell_birnen <= bestell_birnen:
    print(bestell_birnen, 'Birnen                      EUR', "%.2f" % summe_birnen)
else: 
    bestell_birnen > birnen_lager
    print('Tut mir Leid es gibt nich genug Birnen in Lager')

bestell_bananen= int(input('Wieviel Bananen hätten sie gern'))
summe_bananen = bestell_bananen * preis_bananen
if bestell_bananen <= bestell_bananen:
    print(bestell_bananen, 'Bananen                     EUR', "%.2f" % summe_bananen)
    
else: 
    bestell_bananen > bananen_lager
    print('Tut mir Leid es gibt nich genug Bananen in Lager')


summe = ((summe_aepfel)+(summe_birnen)+(summe_bananen))

print(' ')
print('*****************************************')
print('Ihre Gesamtbetrag ist:        EUR', "%.2f" % summe)

print('*****************************************')
print('')
geld = float(input('Mit wie viel Geld werden sie bezahlen: EUR ' ))

rueckgeld = geld - summe

if rueckgeld>=summe:
    print('Sie bekommen EUR', "%.2f" % rueckgeld, ' Rückgeld')
    print(' ')
    print('Vielen Dank für ihre Einkauf')
else:
    print('es fehlen noch', )
    print('Ihre Gesamtbetrag ist:        EUR', "%.2f" % summe)
```
